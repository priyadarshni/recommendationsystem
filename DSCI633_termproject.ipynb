```python
import pandas as pd
import numpy as np
import io
import requests
url="https://raw.githubusercontent.com/travisdesell/exact/master/datasets/2018_coal/burner_10.csv"
s=requests.get(url).content
df=pd.read_csv(io.StringIO(s.decode('utf-8')))
df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
item_similarity=cosine_similarity(df.T)
print(item_similarity)
item_similarity_df=pd.DataFrame(item_similarity,index=df.columns,columns=df.columns)
item_similarity_df
```

```python
def get_similar_attributes(attr, rating):
    score=item_similarity_df[attr]*(rating-2.5)
    score=score.sort_values(ascending=False)
    
    return score
print(get_similar_attributes("Total_Comb_Air_Flow",5))
```

```python
recommend_param= [("Total_Comb_Air_Flow",5),("Primary_Air_Flow",4)]
similar_param=pd.DataFrame()
for param,ratings in recommend_param:
    similar_param= similar_param.append(get_similar_attributes(param,ratings),ignore_index = True)
similar_param.head()
similar_param.sum().sort_values(ascending=False)
```

```python
#final for new cosine matrix
import pandas as pd
import numpy as np
df1 = pd.read_csv("C:/Users/Priya/Desktop/ProjectDSCI/cosine_sim.csv")
df1
```

```python
from scipy.sparse import csr_matrix
#features_df_matrix= csr_matrix(df1.values)
from sklearn.neighbors import NearestNeighbors
model_knn=NearestNeighbors(metric='cosine',algorithm='brute')
#model_knn.fit(df1)
pivot_mat=df1.pivot_table(columns='Burner')
pivot_mat
model_knn.fit(pivot_mat)
```

```python
pivot_mat.shape
```

```python
query_index=np.random.choice(pivot_mat.shape[0])
print(query_index)
distances, indices=model_knn.kneighbors(pivot_mat.iloc[query_index,:].values.reshape(1,-1),n_neighbors=3)
```

```python
pivot_mat.head()
```

```python
for i in range(0, len(distances,flatten())):
    if i==0:
        print("Recommendations for (0):\n",format(pivot_mat.index[query_index]))
        else: 
        print("(0):(1) with distance of (2):",format(i,pivot_mat.index[indices.flatten()[1]],distances.flatten()[1])
```
